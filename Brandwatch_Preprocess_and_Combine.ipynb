## Bring in the Brandwatch Data sets one at a time (2018/2019/2020) and then combine

```python
# load libraries
import pandas as pd
import numpy as np
import keras
```

```python
#Load in the 2018 Brandwatch Data
BW_2018 = pd.read_csv("Brandwatch - 2018.csv", engine='python', header=None)
BW_2018 = BW_2018[[2,8,16,22,42,56]]
BW_2018.rename(columns = {2:'Brand', 8:'ID',16:'Date',22:'Review',42:'Sentiment', 56:'Source'}, inplace = True)
BW_2018.info()
BW_2018.head()
```

```python
#Load in the 2019 Brandwatch Data
BW_2019 = pd.read_csv("Brandwatch - 2019.csv", engine='python', header=None)
BW_2019 = BW_2019[[2, 8,16,22,42,56]]
BW_2019.rename(columns = {2:'Brand', 8:'ID',16:'Date',22:'Review',42:'Sentiment', 56:'Source'}, inplace = True)
BW_2019.info()
BW_2019.head()
```

```python
#Load in the 2020 Brandwatch Data
BW_2020 = pd.read_csv("Brandwatch - 2020.csv", engine='python', header=None)
BW_2020 = BW_2020[[2,8,16,22,42,56]]
BW_2020.rename(columns = {2:'Brand', 8:'ID',16:'Date',22:'Review',42:'Sentiment', 56:'Source'}, inplace = True)
BW_2020.info()
BW_2020.head()
```

```python
# Combine all three years' data
Brandwatch_All = pd.concat([BW_2018, BW_2019, BW_2020],axis=0)
Brandwatch_All.info()
Brandwatch_All["Brand"].value_counts()
```

```python
#Export comnined Brandwatch to csv
Brandwatch_All.to_csv('Brandwatch_All.csv')
```

```python
# ID Retweet Reviews and create new df that does not contain these duplicates. 
Brandwatch_All.Review=Brandwatch_All.Review.astype(str)
Brandwatch_noRT = Brandwatch_All[~Brandwatch_All.Review.str.contains("RT", na=False)]
Brandwatch_noRT.head(10)
```

```python
# removes roughly 50k redundant documents
Brandwatch_noRT.info()
```

```python
#Export this to csv
Brandwatch_noRT.to_csv('Brandwatch_All_noRT.csv')
```
